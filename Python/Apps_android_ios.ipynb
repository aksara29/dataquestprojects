```python
# Define 'explore_data' function
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
```

```python
# Create a non-English detection function
def detect(a):
    i = 0
    
    for chr in a:
            if ord(chr) > 127:
                i = i + 1
                
    if i > 3:
        return False
    else:
        return True
```

```python
print(detect('Docs To Go™ Free Office Suite'))
print(detect('爱奇艺PPS -《欢乐颂2》电视剧热播'))
```

```python
# Create a list of data that removes non-English apps
android_english = []

for row in android_clean:
    name = row[0]
    if detect(name):
        android_english.append(row)
        
ios_english = []

for row in ios[1:]:
    name = row[1]
    if detect(name):
        ios_english.append(row)
        
explore_data(android_english,0,5, True)
print('\n')
explore_data(ios_english,0,5, True)
```

So far, we have cleaned our data: duplicates, non-English apps, free app filter. Our goal is to determine apps that are potentially drawing more users.

Our assumptions are:
1. Build a minimal Android version of the app, and add it to Google Play.
2. If the app has a good response from users, we develop it further.
3. If the app is profitable after six months, we build an iOS version of the app and add it to the App Store.

```python
# Build a dictionary of genre frequency

ios_genres = {}
android_genres = {} 

# iOS
for v in ios_free_apps:
    genres = v[11]
    total = len(ios_free_apps)
    
    if genres in ios_genres:
        ios_genres[genres] += (1/total)*100
    else:
        ios_genres[genres] = (1/total)*100

# Android
for v in android_free_apps:
    genres = v[1]
    total = len(android_free_apps)
    
    if genres in android_genres:
        android_genres[genres] += (1/total)*100
    else:
        android_genres[genres] = (1/total)*100
```

```python
# We want to sort the data for better visualiation

# Create a sorted data list: android
android_table = []
for v in android_genres:
    tuple = (android_genres[v], v) #value, key
    android_table.append(tuple)

table_sorted = sorted(android_table, reverse = True)
for entry in table_sorted:
    print(entry[1], ':', entry[0])
    
# Create a sorted data list: iOS
ios_table = []
for v in ios_genres:
    tuple = (ios_genres[v], v)
    ios_table.append(tuple)

table_sorted_2 = sorted(ios_table, reverse = True)
for entry in table_sorted_2:
    print(entry[1], ':', entry[0])
```

```python
# Alternative way
# Create funtions to a) generate freq tables (percentages) and b) sorting
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        col = row[index]
        if col in table:
            table[col] += 1
        else:
            table[col] = 1
    
    table_percent = {}
    for key in table:
        table_percent[key] = (table[key] / total)*100
        
    return table_percent
```

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key) #value, key as tuple
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
display_table(ios_free_apps, -5)
```

```python
table_no_ratings_by_genre = []
for genre in freq_table(ios_free_apps,-5):
    total = 0
    len_genre = 0
    for app in ios_free_apps:
        genre_app = app[-5]
        if genre_app == genre: #vlookup
            no_ratings = float(app[5])
            total += no_ratings
            len_genre += 1
   
    avg_no_ratings = total / len_genre
    table_no_ratings_by_genre.append([avg_no_ratings, genre])
    
```
